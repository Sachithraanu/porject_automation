```python
import pandas as pd
```

```python
df = pd.read_csv('jan_toyota_1.csv')
df
```

```python
df1 = df.reset_index()
df1 = df.dropna()
df1 = df1.drop_duplicates(subset='Ref', keep="last")
df1.insert(0, 'ID', range(1, 1 + len(df1)))
df1['Year'] = df1['Model'].str[:4]
df1['Full_Name'] = df1['Model'].str[4:]
df1
```
